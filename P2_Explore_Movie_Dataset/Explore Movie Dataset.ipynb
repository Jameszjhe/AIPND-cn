**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
nona_data['profit'] = nona_data['revenue_adj'] - nona_data['budget_adj']
profit = nona_data.groupby('release_year').agg({'profit':'mean'}).sort_values(by=['release_year'], ascending=True)
revenue = nona_data.groupby('release_year').agg({'revenue_adj':'mean'}).sort_values(by=['release_year'], ascending=True)
budget = nona_data.groupby('release_year').agg({'budget_adj':'mean'}).sort_values(by=['release_year'], ascending=True)

plt.plot(revenue, label="Revenue")
plt.plot(profit, label="Profit")
plt.plot(budget, label="Budget")

plt.xlabel('Year')

plt.legend()
plt.grid()
plt.show()
```

从上图可知，电影行业净利润总体呈现下降趋势。电影制作成本90年代后期达到顶峰，自2000年后逐渐下降。票房收入从2000年开始逐渐下降。

**[选做]任务3.3：**选择最多产的10位导演（电影数量最多的），绘制他们排行前3的三部电影的票房情况，并简要进行分析。

```python
directors = nona_data.groupby('director').size().sort_values(ascending=False)
top_10_directors = directors.drop('').head(10).index

values = {}

for index in top_10_directors:
    data = nona_data[nona_data['director'] == index]
    data = data.sort_values(by=['revenue'], ascending=False).head(3)
    values[index] = [x for x in data['revenue'].values]

top_movies = pd.DataFrame(values)

top_1 = top_movies.iloc[0]
top_2 = top_movies.iloc[1]
top_3 = top_movies.iloc[2]

plt.bar(top_10_directors, top_1, label = "Top 1")
plt.bar(top_10_directors, top_2, label = "Top 2", bottom = top_1)
plt.bar(top_10_directors, top_3, label = "Top 3", bottom = top_2 + top_1)

plt.xticks(rotation=80)
plt.legend()
plt.ylabel('Revenue')
plt.xlabel('Director')

plt.show()
```

Steven Sphielberg的票房遥遥领先。Ridley Scott和Ron Howard票房不相伯仲。Woody Allen票房最少。

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
nona_data['release_date'] = pd.to_datetime(nona_data['release_date'], infer_datetime_format=True)
nona_data['release_date'] = nona_data['release_date'].mask(nona_data['release_date'].dt.year > 2018, nona_data['release_date'] - pd.Timedelta(100, unit='Y'))

movie_data_1968_to_2015 = nona_data[(nona_data['release_date'].dt.year >= 1968) & (nona_data['release_date'].dt.year <= 2015) & (nona_data['release_date'].dt.month == 6)]

tmp_data = movie_data_1968_to_2015[['release_date', 'original_title']].groupby(pd.Grouper(key='release_date', freq='M')).count()
tmp_data = tmp_data.loc[tmp_data.index.month == 6]

plt.figure(figsize=(15, 4))
plt.plot(tmp_data.index, tmp_data.values)
plt.xticks(tmp_data.index, rotation=80)
plt.xlabel('Year')
plt.ylabel('The number of movies')
plt.title('The number of movies in June')
plt.show()
```

电影数量逐年增加，2000年后增速增大。

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
comedy_data = movie_data_1968_to_2015[movie_data_1968_to_2015['genres'].str.contains('Comedy', na=False)]
comedy_data = comedy_data[['release_date', 'genres']].groupby(pd.Grouper(key='release_date', freq='M')).count()
comedy_data = comedy_data.loc[comedy_data.index.month == 6]

drama_data = movie_data_1968_to_2015[movie_data_1968_to_2015['genres'].str.contains('Drama', na=False)]
drama_data = drama_data[['release_date', 'genres']].groupby(pd.Grouper(key='release_date', freq='M')).count()
drama_data = drama_data.loc[drama_data.index.month == 6]

plt.figure(figsize=(15, 4))
plt.plot(comedy_data.index, comedy_data.values, label="Comedy")
plt.plot(drama_data.index, drama_data.values, label="Drama")
plt.xticks(tmp_data.index, rotation=80)
plt.xlabel('Year')
plt.ylabel('The number of movies')
plt.title('The number of movies in June')
plt.legend()
plt.show()
```

Comedy与Drama数量呈现逐渐上升的趋势。
